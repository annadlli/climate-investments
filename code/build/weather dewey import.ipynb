```python
#filter further
import polars as pl

INP = "/Users/anna/Desktop/climate-investments/data/attomfire_combined.parquet"
OUT = "/Users/anna/Desktop/climate-investments/data/attomfire_combined_ownerocc_Y_sfu.parquet"

lf = pl.scan_parquet(INP)
present = set(lf.collect_schema().names())

# single-unit family homes
if "UNITSCOUNT" in present:
    lf = lf.filter(
        pl.col("UNITSCOUNT").cast(pl.Int64, strict=False) == 1
    )

# owner-occupied == "Y" only
if "STATUSOWNEROCCUPIEDFLAG" in present:
    lf = lf.filter(
        pl.col("STATUSOWNEROCCUPIEDFLAG")
        .cast(pl.Utf8, strict=False)
        .str.strip_chars()
        .str.to_uppercase() == "Y"
    )

# write filtered parquet (streaming, low memory)
lf.sink_parquet(OUT, compression="zstd")

print(f"✅ Wrote filtered parquet:\n{OUT}")
```

```python
# aggregate for visuals (ZIP x year tenure) - NO sentinel handling
import polars as pl
import os

P = "/Users/anna/Desktop/climate-investments/data/attomfire_combined_ownerocc_Y_sfu.parquet"
OUT_AGG = "/Users/anna/Desktop/climate-investments/data/attom_tenure_zip_year.parquet"
os.makedirs(os.path.dirname(OUT_AGG), exist_ok=True)

lf = pl.scan_parquet(P)

# --- keep minimal columns only ---
keep = [
    "PROPERTYADDRESSZIP",
    "ASSESSORHISTORYYEAR",
    "TAXYEARASSESSED",
    "ASSESSORLASTSALEDATE",          # date
    "LASTOWNERSHIPTRANSFERDATE",     # date
]
present = set(lf.collect_schema().names())
keep = [c for c in keep if c in present]
lf = lf.select(keep)

# ---- build everything in ONE with_columns call ----
lf = lf.with_columns([
    # zip clean
    pl.col("PROPERTYADDRESSZIP")
      .cast(pl.Utf8, strict=False)
      .str.replace_all(r"[^0-9]", "")
      .str.zfill(5)
      .alias("zip"),

    # analysis year (prefer assessor history year)
    pl.coalesce([
        pl.col("ASSESSORHISTORYYEAR").cast(pl.Int32, strict=False),
        pl.col("TAXYEARASSESSED").cast(pl.Int32, strict=False),
    ]).alias("year"),

    # sale year from date columns (no sentinel treatment)
    pl.coalesce([
        pl.col("ASSESSORLASTSALEDATE").dt.year(),
        pl.col("LASTOWNERSHIPTRANSFERDATE").dt.year(),
    ]).cast(pl.Int32, strict=False).alias("sale_year"),
])

# tenure + filters
lf = (
    lf.with_columns((pl.col("year") - pl.col("sale_year")).alias("tenure"))
      .filter(
          pl.col("zip").is_not_null() &
          (pl.col("zip") != "") &
          pl.col("year").is_not_null() &
          pl.col("sale_year").is_not_null() &
          (pl.col("tenure") >= 0) &
          (pl.col("tenure") <= 120)   # allow long tenure since you keep 1900
      )
)

# quick sanity check (safe)
check = lf.select([
    pl.len().alias("rows_kept"),
    pl.col("tenure").min().alias("tenure_min"),
    pl.col("tenure").max().alias("tenure_max"),
]).collect()
print(check)

# aggregate
agg = (
    lf.group_by(["zip", "year"])
      .agg(
          avg_tenure=pl.col("tenure").mean(),
          med_tenure=pl.col("tenure").median(),
          n_props=pl.len()
      )
)

agg.sink_parquet(OUT_AGG, compression="zstd")
print("✅ wrote:", OUT_AGG)
```
